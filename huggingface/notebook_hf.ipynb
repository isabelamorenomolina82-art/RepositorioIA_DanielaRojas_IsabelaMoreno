```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")

textos = [
    "Estoy muy contento con la clase de hoy.",
    "No entiendo nada y me siento frustrado.",
    "La materia está regular."
]

for t in textos:
    print(t, "→", sentiment(t))
```

```python
from transformers import pipeline
from PIL import Image

# Cargar modelo de clasificación de imágenes
classifier = pipeline("image-classification", model="google/vit-base-patch16-224")

# Abrir la imagen que subiste a Colab
img = Image.open("/content/photo.jpg")

# Hacer predicciones
preds = classifier(img)
preds[:5]
```
